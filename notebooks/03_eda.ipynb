```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("default")
```

```python
df = pd.read_csv("../data/WA_Fn-UseC_-Telco-Customer-Churn.csv")
```

```python
df["TotalCharges"] = pd.to_numeric(
    df["TotalCharges"],
    errors="coerce"
)

df["TotalCharges"] = df["TotalCharges"].fillna(0)
```

```python
plt.figure(figsize=(6,4))

sns.countplot(
    x="Churn",
    data=df
)

plt.title("Customer Churn Distribution")

plt.show()
```

```python
plt.figure(figsize=(6,4))

sns.countplot(
    x="gender",
    hue="Churn",
    data=df
)

plt.title("Gender vs Churn")

plt.show()
```

```python
pd.crosstab(
    df["gender"],
    df["Churn"]
)
```

```python
pd.crosstab(
    df["SeniorCitizen"],
    df["Churn"]
)
```

```python
plt.figure(figsize=(6,4))

sns.countplot(
    x="SeniorCitizen",
    hue="Churn",
    data=df
)

plt.title("Senior Citizen vs Churn")

plt.show()
```

```python
pd.crosstab(
    df["Contract"],
    df["Churn"]
)
```

```python
plt.figure(figsize=(8,5))

sns.countplot(
    x="Contract",
    hue="Churn",
    data=df
)

plt.title("Contract Type vs Churn")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    x="Churn",
    y="tenure",
    data=df
)

plt.title("Churn vs Tenure")

plt.show()
```

```python
df.groupby("Churn")["MonthlyCharges"].describe()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    x="Churn",
    y="MonthlyCharges",
    data=df
)

plt.title("Monthly Charges vs Churn")
plt.xlabel("Churn")
plt.ylabel("Monthly Charges")

plt.show()
```

```python
df.groupby("Churn")["TotalCharges"].describe()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    x="Churn",
    y="TotalCharges",
    data=df
)

plt.title("Total Charges vs Churn")
plt.xlabel("Churn")
plt.ylabel("Total Charges")

plt.show()
```

```python
pd.crosstab(
    df["InternetService"],
    df["Churn"],
    normalize="index"
)*100
```
